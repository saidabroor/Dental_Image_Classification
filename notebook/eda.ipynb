```python
import os
import matplotlib.pyplot as plt

healthy_count = len(os.listdir('../data/Healthy Teeth'))
caries_count = len(os.listdir('../data/Caries'))

plt.bar(['Healthy', 'Caries'], [healthy_count, caries_count])
plt.title('Class Distribution')
plt.ylabel('Number of images')
plt.show()
```

```python
from PIL import Image

for img_name in os.listdir(r'data/Healthy Teeth')[:5]:
    img = Image.open(f'data/Healthy Teeth/{img_name}')
    print(img_name, img.size)
```

```python
data_folder = 'data/Healthy Teeth'

for img_name in os.listdir(data_folder)[:3]:
    img_path = os.path.join(data_folder, img_name)
    img = Image.open(img_path)
    print(img_name, img.size)
    
    plt.imshow(img)
    plt.axis('off')  # Hide axes
    plt.title(img_name)
    plt.show()
```

```python
import cv2

fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for i, img_name in enumerate(os.listdir('data/Healthy Teeth')[:5]):
    img = cv2.imread(f'data/Healthy Teeth/{img_name}')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title("Healthy")
    axes[i].axis('off')

plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 5))

for i, img_name in enumerate(os.listdir('data/Caries')[:5]):
    img = cv2.imread(f'data/Caries/{img_name}')
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title("Healthy")
    axes[i].axis('off')

plt.show()
```
